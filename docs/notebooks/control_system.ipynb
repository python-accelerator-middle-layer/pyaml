# Control System set up

The control system interface is in a separated package from pyaml

pip install accelerator-middle-layer   #  <-- pyaml itself
pip install tango-pyaml                #  <-- tango interface to pyaml 
pip install pyaml-oa-cs                #  <-- ophyd-asynch interface to pyaml

If the above packages are installed the examples below will work (also in the same script)

````python
from pyaml.accelerator import Accelerator
````

Assuming these lines are present in ../../tests/config/EBSTune_TANGO.yaml
```
type: pyaml.accelerator
name: sr
energy: 6e9
controls:
  - type: tango.pyaml.controlsystem
    tango_host: ebs-simu-3:10000
    name: live
...
```

````python
# work with TANGO
SR_Tango: Accelerator = Accelerator.load('../../tests/config/EBSTune_TANGO.yaml')
````

Assuming these lines are present in ../../tests/config/EBSTune_OPHYDASYNCH.yaml
```
type: pyaml.accelerator
name: sr
energy: 6e9
controls:
  - type: pyaml-cs-oa.pyaml.controlsystem
    name: live
...
```

````python
# work with Ophyd-Asynch
SR_OphydAsynch: Accelerator = Accelerator.load('../../tests/config/EBSTune_OPHYDASYNCH.yaml')
````
