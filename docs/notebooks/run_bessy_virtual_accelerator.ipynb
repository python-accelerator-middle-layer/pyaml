Instruction to run example scripts using the BESSYII virtual accelerator

instruction to install apptainer (and run the bessy virutal accelerator)

if ARM-based macos (M1, M2, M3....)
server:
1) brew install qemu lima
2) limactl start --rosetta --vm-type=vz --network=vzNAT template://apptainer
3) press enter

if Intel-based macos
server:
1) brew install lima
2) limactl start --vm-type=vz --network=vzNAT template://apptainer
3) press enter

To start container:
1) limactl shell apptainer 
2) apptainer run oras://registry.hzdr.de/digital-twins-for-accelerators/containers/pyat-softioc-digital-twin:v0-1-2-bessy.2475331


Stopping container:
1) limactl stop apptainer
2) limactl delete apptainer


to get address we need pvlist. On another terminal:
1) limactl shell apptainer # to go in VM
2) apptainer run oras://registry.hzdr.de/digital-twins-for-accelerators/epics-tools:v0-1-0.2028728`
3) pvlist
4) pvlist 192.168.5.15:5075 (replace the address with the one printed by the command above after tcp@)

create an environment

python -m venv bessy2_pyaml
source bessy2_pyaml/bin/activate

clone repositories from main (not latest release)

git clone https://github.com/python-accelerator-middle-layer/pyaml.git
git clone https://github.com/python-accelerator-middle-layer/pyaml-cs-oa.git

install locally (need to modify the pyaml configuration)

cd pyaml
pip install -e . 

cd pyaml-cs-oa
pip install -e . 

pip install aioca

change the pyaml configuration file ./pyaml/examples/BESSY2_example/BESSY2Orbit.yaml at line 11 with the correct prefix (replace pons: by your prefix)
change the pyaml configuration file ./pyaml/examples/BESSY2_example/BESSY2Tune.yaml at line 11 with the correct prefix (replace pons: by your prefix)
change the pyaml configuration file ./pyaml/examples/BESSY2_example/BESSY2Chroma.yaml at line 11 with the correct prefix (replace pons: by your prefix)


navigate to the folder ./pyaml/examples/BESSY2_example/

run the file bessy2-orbit.py
(this will take sometime to compute the ORM)

run the file bessy2-tune.py
run the file bessy2-chroma.py
