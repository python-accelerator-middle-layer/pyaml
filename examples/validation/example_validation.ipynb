```python
# Create the schema registry and discover schemas
# This should for the users be called automatically

from pyaml.validation import SchemaRegistry

schema_registry = SchemaRegistry()
schema_registry.discover()
```

```python
# Print the content of the schema registry

schema_registry
```

```python
# Read in a configuration file

import yaml

with open("../BESSY2_example/BESSY2Tune_new_format.yaml", "r") as f:
    config = yaml.safe_load(f)
```

```python
# Validate the data

from pyaml.validation import SchemaValidator

validated = SchemaValidator.validate(config)
```

```python
# Print the validated data

validated.model_dump()
```
